## HANDLING MISSING VALUES 

```python
#Drops columns with a percentage of missing values exceeding the specified threshold.Parameters:df (DataFrame): The input DataFrame.
#threshold (float): The threshold for the percentage of missing values (default is 0.7).Returns:DataFrame: A new DataFrame with high-missing 
#columns removed.
def drop_high_missing_columns(f, threshold=0.7):
    
    total_rows = f.shape[0]
    missing_threshold = threshold * total_rows

    # Calculate the number of missing values in each column
    missing_counts = f.isnull().sum()

    # Identify columns with missing values exceeding the threshold
    high_missing_columns = missing_counts[missing_counts > missing_threshold].index

    # Drop the high-missing columns
    df_cleaned = f.drop(columns=high_missing_columns)

    return df_cleaned
```

```python
#We're dropping the columns with more than 70% of missing values.
#It helps streamline your dataset by eliminating columns that may not contribute meaningful information and could potentially introduce 
#noise into your analysis. Removing such columns can lead to a more focused and efficient modeling process.
df1 = drop_high_missing_columns(df, threshold=0.7)
```

```python
from sklearn.decomposition import TruncatedSVD

# Define the size of the random sample
sample_size = 10000  # Adjust this value as needed



# Create a random sample of your dataset
sampled_df = df1.sample(n=sample_size, random_state=42)

# Separate the features and target variable
features = sampled_df.drop(columns=['PRIM_CONTRIBUTORY_CAUSE'])
target = sampled_df['PRIM_CONTRIBUTORY_CAUSE']

# Define categorical and numerical features
categorical_features = ['TRAFFIC_CONTROL_DEVICE', 'WEATHER_CONDITION', 'LIGHTING_CONDITION', 'FIRST_CRASH_TYPE', 'TRAFFICWAY_TYPE', 'ALIGNMENT', 'ROADWAY_SURFACE_COND', 'ROAD_DEFECT', 'MOST_SEVERE_INJURY', 'INJURIES_FATAL', 'INJURIES_INCAPACITATING', 'INJURIES_NON_INCAPACITATING', 'INJURIES_REPORTED_NOT_EVIDENT', 'INJURIES_NO_INDICATION', 'INJURIES_UNKNOWN', 'CRASH_HOUR', 'CRASH_DAY_OF_WEEK', 'CRASH_MONTH']
numeric_features = ['POSTED_SPEED_LIMIT', 'STREET_NO', 'NUM_UNITS']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Modify your preprocessing pipeline to include TruncatedSVD for dimensionality reduction
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),  # Standardize numerical features
        ('cat', OneHotEncoder(drop='first', handle_unknown='ignore'), categorical_features)  # One-hot encode categorical features
    ])

# Include TruncatedSVD as a dimensionality reduction step
n_components = 10  # Adjust the number of components as needed
tsvd = TruncatedSVD(n_components=n_components)

# Create an RFE estimator
num_features_to_select = 25
rfe_estimator = RFE(estimator=RandomForestClassifier(), n_features_to_select=num_features_to_select)

# Modify the full pipeline to include TruncatedSVD, RFE, and the classifier
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('tsvd', tsvd),  # Include TruncatedSVD for dimensionality reduction
    ('feature_selection', rfe_estimator)  # Include RFE step
])
```

```python
# Fit the pipeline to the training data
pipeline.fit(X_train, y_train)
# Transform the training data
X_train_transformed = pipeline.transform(X_train)
# Transform the test data using the same pipeline
X_test_transformed = pipeline.transform(X_test)
```
